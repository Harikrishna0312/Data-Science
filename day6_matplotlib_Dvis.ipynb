### a.Line chart

```python
year = [1950,1960,1970,1980,1990,2000,2010,2020]
gdp =[300,540,670,980,1020,1100,1300,1478]
```

```python
plt.figure(figsize=(6,3))
plt.title('GDP across the decades')
plt.xlabel('year')
plt.ylabel('GDP')

#Plot the line
plt.plot(year, gdp,
        marker='H', markersize=10, markerfacecolor='orange', markeredgecolor='midnightblue',
        linestyle='dotted', color='midnightblue'
        )

plt.show()
```

#### b. Scatter Plot

```python
prices=pd.read_csv("C:/Users/harik/python pratice/house_prices.csv")
prices.head()
```

```python
prices.rename(columns={
    'Area (ft.)':'Area'
}, inplace=True)
```

```python
prices.head()
```

```python
prices['Building Type'].value_counts()
```

```python
plt.figure(figsize=(10,3))
plt.title('House Prices')
plt.xlabel('Area (in sq.ft)')
plt.ylabel('Price')

plt.scatter(prices['Area'], prices['Price'],
           c='orange', edgecolor='midnightblue',
            alpha=0.6, 
            s=50
           )
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.title('House Prices')
plt.xlabel('Area (in sq.ft)')
plt.ylabel('Price')

plt.scatter(prices['Area'], prices['Price'],
           c=prices['Year of sale'], edgecolor='midnightblue',
            alpha=0.6, 
            s=prices['Building Type']*50,
            cmap='BuPu'
           )
plt.show()
```

#### 1.Bar Chart

```python
cars=pd.read_csv("C:/Users/harik/python pratice/bar_chart_data.csv")
cars
```

```python
plt.figure(figsize=(10,3))
plt.title('Used Cars')
plt.xlabel('Brand', color='#F5276C')
plt.ylabel('no of Listings', color='#F87C63')
plt.xticks(fontsize=8, rotation=20)
plt.ylim(0,1000)

### Bar Chart
ax=plt.bar(cars['Brand'],cars['Cars Listings'], color='#FCC6BB',
       alpha=0.8,
       edgecolor='black'
       )
plt.bar_label(ax)
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.title('Used Cars')
plt.ylabel('Brand', color='#F5276C')
plt.xlabel('no of Listings', color='#F87C63')
plt.xticks(fontsize=8, rotation=15)

### Bar Chart
cars_sort=cars.sort_values(by='Cars Listings')
ax=plt.barh(cars_sort['Brand'],cars_sort['Cars Listings'], color='teal',
       alpha=0.8,
       edgecolor='black'
       )
plt.bar_label(ax)
plt.show()
```

#### 2.Pie Chart

```python
courses=['Data Science', 'Analytics', 'PowerBi', 'Tableau', 'Mysql']
students=[128,205,178,60,110]
```

```python
plt.pie(students, labels=courses,
        autopct= '%0.1f%%',
        explode=[0,0,0,2,0],
        colors=['orange','cyan','pink','tomato','#FEF3C6']
       )
plt.show()
```
