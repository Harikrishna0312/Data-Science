### <font color='Teal'> Data Cleaning </font>

#### <font color='Teal'> Handle Missing Values </font>

```python
dnull = [ [1,2,3,'M',7],
         [np.nan,5,6,np.nan,7],
         [8,np.nan,9,'F',10],
         [np.nan,np.nan,np.nan,'M',11] 
        ]

dframe = pd.DataFrame(dnull, columns=['a','b','c','d','e'])
dframe
```

```python
dframe.info()
```

```python
dframe.isna()
```

```python
dframe.isna().sum()
```

#### Drop null values

```python
# 1. Remove rows with null values
dframe.dropna(axis=0,
    #inplace=True
)
```

```python
dframe
```

```python
dframe.dropna(axis=1,
    #inplace=True
)
```

```python
avg_a = dframe['a'].mean()
avg_b = dframe['b'].mean()
avg_c = dframe['c'].mean()

avg_a, avg_b, avg_c
```

```python
dframe['d'].mode()
```

```python
mode_d = dframe['d'].mode()[0]
mode_d
```

```python
# 2. Replace nr values with mean/median
# Replace categorical values with mode
dframe.fillna(
  { 'a' : avg_a,
    'b' : avg_b,
    'c' : avg_c,
    'd' : mode_d }, 
    #inplace=True
)
```

```python
# 3. Replace all nulls with the same value
dframe.fillna(0,
              #inplace=True
             )
```

#### merge

```python
d1 = {'movieid':[1,2,3,4,5],
     'title':['3 Idiots','2 States','Leo','RRR','Pushpa']}
movies = pd.DataFrame(d1)
movies
```

```python
d2 = {
    'memid':[1,2,3,4],
    'name':['Tom','Tim','Jon','Jim'],
    'movieid':[1,np.nan,3,5]
}
members=pd.DataFrame(d2)
members
```

```python
pd.merge(movies,members, on='movieid', how='inner')
```

```python
pd.merge(movies,members, on='movieid', how='left')
```

```python
pd.merge(movies,members, on='movieid', how='right')
```

```python
pd.merge(movies,members, on='movieid', how='outer')
```

```python
d3 = {'id':[1,2,3,4,5],
     'title':['3 Idiots','2 States','Leo','RRR','Pushpa']}
movies2 = pd.DataFrame(d3)
movies2
```

```python
pd.merge(movies2,members, left_on='id', right_on='movieid', how='inner')
```

#### apply() - applies a funtion on every row in a series

```python
def efn(x):
    if x > 4000:
        res = 'Good Salary'
    else:
        res = 'Junk Salary'
    return res
```
