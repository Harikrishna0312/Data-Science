#### Loops in python

```python
mark_list = [34,56,77,89,76,45]

for mark in mark_list:
    if mark >= 80 :
        res = 'Distinction'
    elif mark >= 60 :
        res = 'First Class'
    elif mark >= 40:
        res = 'Second Class'
    else:
        res = 'Failed'
    print('The grade is :', res)
```

```python
def grade_of(mark):
    if mark >= 80 :
        res = 'Distinction'
    elif mark >= 60 :
        res = 'First Class'
    elif mark >= 40:
        res = 'Second Class'
    else:
        res = 'Failed'
    return res
```

```python
def pyramid(rows):
    '''
    Prints a pyramid of stars
    Parameter: Integer representing no of rows
    Returns : None
    '''
    for i in range(1, rows + 1):
        print(' ' * (rows - i) + '*' * (2 * i - 1))
```

```python
pyramid(7)
```

```python
def grade_list(mark_list):
    '''
    Parameter: A list of numbers
    Returns: A list of grades
    '''
    grade_list = []
    for mark in mark_list:
        if mark >= 80 :
            res = 'Distinction'
        elif mark >= 60 :
            res = 'First Class'
        elif mark >= 40:
            res = 'Second Class'
        else:
            res = 'Failed'
        grade_list.append(res)
    return grade_list
```

```python
grade_list([34,56,78,89,90])
```
