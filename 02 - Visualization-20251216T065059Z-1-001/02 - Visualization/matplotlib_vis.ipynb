##### b. Scatter Plot

```python
prices = pd.read_csv("C:/Users/harik/python pratice/00 - datasets-20251216T065029Z-1-001/00 - datasets/house_prices.csv")
prices.head()
```

```python
prices.rename(columns={
    'Area (ft.)':'Area'
}, inplace=True)
```

```python
prices.head()
```

```python
prices['Building Type'].value_counts()
```

```python
plt.figure(figsize=(6,3))
plt.title('House Prices')
plt.xlabel('Area (in Sq.Ft)')
plt.ylabel('Price')

plt.scatter(prices['Area'], prices['Price'],
            c='orange', edgecolor='midnightblue', 
            alpha=0.6, 
            s=50
           )
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.title('House Prices')
plt.xlabel('Area (in Sq.Ft)')
plt.ylabel('Price')

plt.scatter(prices['Area'], prices['Price'],
            c=prices['Year of sale'], edgecolor='midnightblue', 
            alpha=0.6, 
            s=prices['Building Type']*50,
            cmap='BuPu'
           )
plt.show()
```

### 2. Categorical vs Nr

##### 1. Bar Chart

```python
cars = pd.read_csv("C:/Users/harik/python pratice/00 - datasets-20251216T065029Z-1-001/00 - datasets/bar_chart_data.csv")
cars
```

```python
plt.figure(figsize=(7,3))
plt.title('Used Cars')
plt.xlabel('Brand', color='#F5276C')
plt.ylabel('No of Listings', color='#F5276C')
plt.xticks(fontsize=8, rotation=15)
plt.ylim(0,1000)


### Bar Chart
ax = plt.bar(cars['Brand'],cars['Cars Listings'], color='#F5276C', 
        alpha=0.8,
        edgecolor='black'
       )
plt.bar_label(ax)
plt.show()
```

```python
plt.figure(figsize=(7,3))
plt.title('Used Cars')
plt.ylabel('Brand', color='teal')
plt.xlabel('No of Listings', color='teal')
plt.xticks(fontsize=8, rotation=15)


### Bar Chart
cars_sort = cars.sort_values(by='Cars Listings')
ax = plt.barh(cars_sort['Brand'],cars_sort['Cars Listings'], color='teal', 
        alpha=0.8,
        edgecolor='black'
       )
plt.bar_label(ax)
plt.show()
```
