### Boxplot

```python
sns.catplot(tips,x='day',y='total_bill',
           kind='box',
           height=3,
           aspect=2,
            hue='day',
           palette='vlag'
           )
plt.show()
```

```python
sns.catplot(tips,x='day',y='total_bill',
           kind='strip',
           height=3,
           aspect=2,
            hue='day',
           palette='viridis'
           )
plt.show()
```

```python
sns.catplot(tips,x='day',y='total_bill',
           kind='swarm',
           height=3,
           aspect=2,
            hue='day',
           palette='viridis'
           )
plt.show()
```

```python
sns.catplot(tips,x='day',y='total_bill',
           kind='swarm',
           height=3,
           aspect=2,
            hue='time',
           palette='viridis'
           )
plt.show()
```

```python
sns.catplot(tips,x='day',y='total_bill',
           kind='violin',
           height=3,
           aspect=2,
           hue='sex',
           split=True,
           palette='vlag'
           )
plt.show()
```

#### Numeric Vs Numaric

```python
mpg=sns.load_dataset('mpg')
mpg.head(3)
```

```python
mpg['origin'].unique()
```

```python
sns.relplot(mpg, x='model_year', y='mpg',
          kind='line',
           height=3,
           aspect=2,
           #ci=False
           errorbar=('ci', False),
           palette='viridis',
           hue='origin',
           style='origin'
          )
plt.show()
```

#### Scatter Plot

```python
hprice=pd.read_csv("C:/Users/harik/python pratice/kc_house_data.csv")
hprice=hprice.head(150)
```

```python
hprice.head(3)
```

```python
sns.relplot(hprice, x='sqft_living', y='price',
          kind='scatter',
           height=3,
           aspect=1.5,
           palette='viridis',
            #hue='origin',
            color='tomato',
           s=100,
            edgecolor='black',
           alpha=0.6
          )
plt.show()
```

```python
sns.relplot(hprice, x='sqft_living', y='price',
          kind='scatter',
           height=4,
           aspect=1.5,
           palette='mako',
            hue='bedrooms',
            #size='sqft_lot',
            s=70,
            edgecolor='black',
           alpha=0.6
          )
plt.show()
```

```python
sns.relplot(tips, x='total_bill', y='tip',
            kind = 'scatter',
            height=3,
            aspect=2,
            s=70,
            hue='sex',
            palette='vlag',
            edgecolor='black'
           )
plt.title('Tip Vs Total Bill Amount')

plt.show()
```

### Univariate Data

```python
sns.displot(hprice, x='price',
           kind='hist',
            #kde=True,
            height=3,
            aspect=1.5,
            color='teal'
           )
plt.show()
```

```python
sns.displot(hprice, x='price',
            kind='kde',
            #kde=True,
            height=3,
            aspect=1.5,
            color='teal'
           )
plt.show()
```
