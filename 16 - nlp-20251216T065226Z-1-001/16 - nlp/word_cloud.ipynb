```python
from PIL import Image
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("modi.txt","r",encoding='utf8') as input:
    text = input.read()
type(text)   
text
```

```python
wordcloud = WordCloud(width=800, height=600, background_color = 'black', stopwords=set(STOPWORDS))
wordcloud.generate(text)
wordcloud.to_file('modi.png')
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

```python
mask_array = np.array(Image.open("C:/Users/harik/OneDrive/Pictures/Camera Roll/itachi 1.jpg"))
wordcloud = WordCloud(width=800, height=600, background_color = 'white', mask=mask_array, stopwords=set(STOPWORDS))
wordcloud.generate(text)
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```
