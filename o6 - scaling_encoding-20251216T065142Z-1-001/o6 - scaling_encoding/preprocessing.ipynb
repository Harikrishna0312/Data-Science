### b. Encoding categorical data

```python
iris = pd.read_csv("C:/Users/harik/python pratice/00 - datasets-20251216T065029Z-1-001/00 - datasets/iris.csv")
iris.head()
```

#### a. Label Encoding

```python
iris1 = iris.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit_transform(iris1['species'])
```

```python
iris1['species'] = le.fit_transform(iris1['species'])
iris1.head()
```

#### b. One Hot Encoding

##### 1. Using OneHotEncoder Class

```python
iris2 = iris.copy()
iris2.head(3)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
enc_data = ohe.fit_transform(iris2[['species']])
enc_data
```

```python
# To display the encoded data
enc_arr = enc_data.toarray()
#enc_arr
```

```python
iris2.join(pd.DataFrame(enc_arr, columns=['Sp_Setosa','Sp_Versicolor','Sp_Virginica']))
```

```python
final_data = iris2.join(pd.DataFrame(enc_arr, columns=['Sp_Setosa','Sp_Versicolor','Sp_Virginica'])).drop('species', axis=1)
final_data
```

#### Drop first

```python
ohe = OneHotEncoder(drop='first')
enc_data = ohe.fit_transform(iris2[['species']])
enc_data.toarray()
```
